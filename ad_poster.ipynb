```python
from manim import *
```

```python
%%manim --fps 60 -v WARNING ad_poster_Scene


bg_color = WHITE
main_color = BLACK

bg_color = DARKER_GRAY
main_color = WHITE


# global manim settings
config.background_color = bg_color
config.frame_size = [2480, 3508]

config.frame_width = 24.80
config.frame_height = 35.08

class ad_poster_Scene(Scene):
	def construct(self):
		# scene geometry
		frame_width = config.frame_width
		frame_height = config.frame_height

		# printing out the scene geometry
		print(f"width of scene:\t\t{frame_width}")
		print(f"height of scene:\t{frame_height}")

		# buffs
		border_buff = 2
		section_buff = 2.25
		bg_rect_buff = 1


		# SECTION: example images
		ex_image_leopard = ImageMobject("media_ad_poster/snow_leopard.jpg", scale_to_resolution = 1080).scale(0.16).align_on_border(LEFT + UP, buff = border_buff)
		ex_image_line_integration = ImageMobject("media_ad_poster/line_integration.jpg", scale_to_resolution = 1080).scale(0.5*0.16).align_on_border(RIGHT + UP, buff = border_buff)

		ex_image_group = Group(ex_image_leopard, ex_image_line_integration)
		self.add(ex_image_group)


		# SECTION: header
		headline = Text("Manim-Hackathon", font_size = 96, weight = BOLD, color = main_color)
		sub_headline = Text("19. - 21. Mrz 2026", font_size = 72, weight = BOLD, color = main_color).next_to(headline, DOWN, buff = 0.75)
		headline_group = VGroup(headline, sub_headline)

		if bg_color == WHITE:
			banner = ManimBanner(dark_theme = False).scale(0.65).next_to(headline_group, LEFT, buff = 0.75)
		else:
			banner = ManimBanner(dark_theme = True).scale(0.65).next_to(headline_group, LEFT, buff = 0.75)

		section_header_group = VGroup(headline, sub_headline, banner).next_to(ex_image_group, DOWN, buff = section_buff)
		self.add(section_header_group)


		# SECTION: logo images
		logo_physics = ImageMobject("media_ad_poster/newtons-cradle_1675238.png", scale_to_resolution = 1080).scale(0.075)
		logo_teams = ImageMobject("media_ad_poster/management_12071897.png", scale_to_resolution = 1080).scale(0.075).next_to(logo_physics, DOWN, buff = 1)
		logo_food = ImageMobject("media_ad_poster/fast-food_1255303.png", scale_to_resolution = 1080).scale(0.075).next_to(logo_teams, DOWN, buff = 1)
		logo_podium = ImageMobject("media_ad_poster/podium_1496240.png", scale_to_resolution = 1080).scale(0.075).next_to(logo_food, DOWN, buff = 1)

		text_physics = Text("Physikalische Animationen", font_size = 54, color = main_color).next_to(logo_physics, RIGHT, buff = 1)
		text_teams = Text("(Python-)Programmierung in Teams", font_size = 54, color = main_color).next_to(logo_teams, RIGHT, buff = 1)
		text_food = Text("Pizza / Getrnke", font_size = 54, color = main_color).next_to(logo_food, RIGHT, buff = 1)
		text_podium = Text("Ehrung bester Teams mit bis zu 50 Preisgeld", font_size = 54, color = main_color).next_to(logo_podium, RIGHT, buff = 1)

		logo_group = Group(logo_physics, logo_teams, logo_food, logo_podium, text_physics, text_teams, text_food, text_podium).next_to(section_header_group, DOWN, buff = section_buff)

		if bg_color == WHITE:
			logo_bg_rectangle_left = BackgroundRectangle(logo_group, color = LIGHTER_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(LEFT, buff = border_buff)
			logo_bg_rectangle_right = BackgroundRectangle(logo_group, color = LIGHTER_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(RIGHT, buff = border_buff)
		else:
			logo_bg_rectangle_left = BackgroundRectangle(logo_group, color = DARK_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(LEFT, buff = border_buff)
			logo_bg_rectangle_right = BackgroundRectangle(logo_group, color = DARK_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(RIGHT, buff = border_buff)

		section_logo_group = Group(logo_bg_rectangle_left, logo_bg_rectangle_right, logo_group).next_to(section_header_group, DOWN, buff = section_buff)
		self.add(section_logo_group)


		# SECTION: QR-Codes
		text_studip = Text("Anmeldung auf StudIP", font_size = 60, weight = BOLD, color = main_color)
		text_info = Text("weitere Informationen", font_size = 60, weight = BOLD, color = main_color)

		arrow_studip = Line(start = [0.5, 0, 0], end = [-0.5, 0, 0], stroke_width = 16, color = RED).next_to(text_studip, LEFT, buff = 0.75).add_tip()
		arrow_info = Line(start = [-0.5, 0, 0], end = [0.5, 0, 0], stroke_width = 16, color = RED).next_to(text_info, RIGHT, buff = 0.75).add_tip()

		studip_group = Group(text_studip, arrow_studip).next_to(section_logo_group, DOWN)
		info_group = Group(text_info, arrow_info).next_to(studip_group, DOWN, buff = 0.5)
		text_mail1 = Text("Meldet euch gerne bei Fragen!", font_size = 32, color = main_color).next_to(info_group, DOWN, buff = 1)
		if bg_color == WHITE: 
			text_mail2 = Text("carlo.voncarnap@stud.uni-goettingen.de", font_size = 32, color = PURE_BLUE).next_to(text_mail1, DOWN, buff = 0.25)
		else:
			text_mail2 = Text("carlo.voncarnap@stud.uni-goettingen.de", font_size = 32, color = BLUE).next_to(text_mail1, DOWN, buff = 0.25)
		qr_text_group = Group(info_group, studip_group, text_mail1, text_mail2)

		qr_info = ImageMobject("media_ad_poster/frame_info_static.png", scale_to_resolution = 1080).scale(0.035).next_to(qr_text_group, RIGHT, buff = 1).align_on_border(RIGHT, buff = border_buff)
		qr_studip = ImageMobject("media_ad_poster/frame_studip_static.png", scale_to_resolution = 1080).scale(0.035).next_to(qr_text_group, LEFT, buff = 1).align_on_border(LEFT, buff = border_buff)
		
		section_qr_group = Group(qr_text_group, qr_info, qr_studip).next_to(section_logo_group, DOWN, buff = section_buff)#.align_on_border(DOWN, buff = 1.5*border_buff)
		self.add(section_qr_group)


		# SECTION: cup
		logo_cup = ImageMobject("media_ad_poster/cup.png", scale_to_resolution = 1080).scale(0.2)
		text_cup = Text("Beste Animation / Beste Physik", font_size = 60, weight = BOLD, color = main_color)
		# if bg_color == WHITE:
		# else:
		# 	text_cup = Text("Beste Animation / Beste Physik", font_size = 60, weight = BOLD, color = YELLOW)
		text_1st = Text("1. Platz: je 50 ", font_size = 54, color = main_color).next_to(text_cup, DOWN, buff = 0.5).shift(3*LEFT)
		text_2nd = Text("2. Platz: je 25 ", font_size = 54, color = main_color).next_to(text_1st, DOWN, buff = 0.5)

		cup_text_group = Group(text_cup, text_1st, text_2nd).next_to(logo_cup, RIGHT, buff = 1.5)
		cup_group = Group(logo_cup, cup_text_group).next_to(qr_text_group, DOWN, buff = section_buff)
		if bg_color == WHITE:
			cup_bg_rectangle_left = BackgroundRectangle(cup_group, color = LIGHTER_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(LEFT, buff = 2)
			cup_bg_rectangle_right = BackgroundRectangle(cup_group, color = LIGHTER_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(RIGHT, buff = 2)
		else:
			cup_bg_rectangle_left = BackgroundRectangle(cup_group, color = DARK_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(LEFT, buff = 2)
			cup_bg_rectangle_right = BackgroundRectangle(cup_group, color = DARK_GRAY, buff = bg_rect_buff, fill_opacity = 1).align_on_border(RIGHT, buff = 2)
		# self.add(cup_bg_rectangle_left, cup_bg_rectangle_right, cup_group)
		# self.add(cup_group)


		CVC = Text('CVC', font_size = 16, weight = BOLD, color = main_color, font = 'Latin Modern Sans').align_on_border(RIGHT + DOWN, buff = 1)
		print(CVC.get_center())
```
