## 0.2 Manim in Jupyter Notebooks

First goal should be to make the following code cell run, it is a simple black frame with a tiny `CVC` in the bottom-right corner. Afterwards the lines are explained.  

````python
%%manim -ql --fps 60 -v WARNING example_Scene


class example_Scene(Scene):
	def construct(self):
		# example code to disply my initials in the bottom-right corner
		CVC = Text('CVC', font_size = 12, weight = BOLD, color = WHITE, font = 'Latin Modern Sans').align_on_border(RIGHT + DOWN, buff = 0.2)
		self.add(CVC)
````

```%%manim -ql --fps 60 -v WARNING example_Scene```

This cell-magic line makes Jupyter render the output directly into the notebook, which is highly convenient for direct feedback and used throughout all these notebooks. The following flags have been used here, more can be found in the [documentation](https://docs.manim.community/en/stable/guides/configuration.html):

* `-ql`: the flag setting the **q**uality of the output. In this case it is set to **l**ow (854x480 15FPS); however, higher resolution can be achieved via `-qm` (1280x720 30FPS), `-qh` (1920x1080 60FPS), `-qp` (2560x1440 60FPS), or `-qk` (3840x2160 60FPS)
* `--fps 60`: here we are setting the framerate to **60FPS** despite the low resolution
* `-v WARNING`: shows warnings I guess
* `example_Scene`: this is the name of the following scene defined by the scene class and **absolutely required**
